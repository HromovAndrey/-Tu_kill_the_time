```python
import torch
from torchvision.io.image import read_image
from torchvision.models.segmentation import fcn_resnet50, FCN_ResNet50_Weights
from torchvision import transforms
```

```python
img = read_image("/content/загружено.png")
```

```python
img = transforms.Resize((520, 520))(img)
```

```python
weights = FCN_ResNet50_Weights.DEFAULT
model = fcn_resnet50(weights=weights)
model.eval()
```

```python
preprocess = weights.transforms(resize_size=None)
```

```python
# Преобразуем изображение в RGB (удаляем альфа-канал, если он есть)
if img.shape[0] == 4:  # Если изображение имеет 4 канала (RGBA)
    img = img[:3, :, :]  # Убираем альфа-канал

# Теперь применяем препроцессинг
batch = preprocess(img)
batch = batch.unsqueeze(0)
```

```python
with torch.no_grad():
  prediction = model(batch)["out"]

masks = prediction.argmax(dim=1)
masks = masks[0]
```

```python
import matplotlib.pyplot as plt

plt.imshow(img.permute(1, 2, 0))
```

```python
plt.imshow(masks)
```

```python
class_to_idx = {cls: idx for (idx, cls) in enumerate(weights.meta["categories"])}
```

```python
class_to_idx
```

```python
from torchvision.utils import draw_segmentation_masks
dog_masks = masks == class_to_idx['dog']
masked_img = draw_segmentation_masks(img, dog_masks, alpha=0.7)
plt.imshow(masked_img.permute(1, 2, 0))
```

```python
img.shape
```

```python
# Клонируем изображение
masked_img = img.clone()

# Преобразуем изображение в Byte, так как исходное изображение имеет тип Byte
masked_img = masked_img.to(torch.uint8)

# Определяем желтый цвет в формате Byte
yellow_pixel = torch.tensor([255, 255, 0], dtype=torch.uint8)

# Присваиваем маску для выделенных пикселей собаки
masked_img[:, dog_masks] = yellow_pixel.view(-1, 1)

# Отображаем результат
plt.imshow(masked_img.permute(1, 2, 0))
```
